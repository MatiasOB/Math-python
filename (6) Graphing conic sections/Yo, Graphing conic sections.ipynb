# Graphing parabolas

Parabolas formula =  $$ 4p(y - k) = (x - h)^2 $$
$$ y = \frac{1}{4p} (x - h)^2 + k $$

```python
# parameters
a = 1  # 1/4p
h = 5  # x-axis offsec
k = -2

n = 100

# x-axos points to evaluate the function

x = np.linspace(h-10, h+10, 100)

# create the function

y = a*(x - h)**2 + k

plt.figure(figsize=(12, 6), dpi=100)

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.grid()


plt.subplot(1, 2, 2)
plt.plot(y, x)
plt.grid()

plt.tight_layout()
```

**The parabola is the set of all points that is equidistant between the focus and the directrix**

```python
# parameters
a = 1  # 1/4p
h = 2  # x-axis offset
k = 1  # y-axis offset
p = a/4

vertex = (h, k)
focus = (h, k + p)
directrix = k - p


n = 100

# x-axos points to evaluate the function

x = np.linspace(h-5, h+5, 100)

# create the function

y = a*(x - h)**2 + k


plt.plot(x, y, label="parabola")
plt.plot(vertex[0], vertex[1], "ro", label="vertex")
plt.plot(focus[0], focus[1], "go", label="focus")
plt.plot(x, [directrix]*len(x), label="directrix")
plt.grid()
plt.ylim([0, 6])
plt.xlim([-3, 6])
plt.legend()
```

# Graphing Circles

General Formula of circles:

$$ (y - k)^2 + (x - h)^2 = r^2 $$

being r = radious, h = location of the center in the x-axis, k = location of the center in the y-axis.<br>
When the center is at the origin the formula is commonly expressed as :

$$ x^2 + y^2 = r^2 $$

```python
h = 2
k = -3
r = 3

axlim = np.max([abs(h),abs(k)]) + r

x = np.linspace(-axlim, axlim,100)

X,Y = np.meshgrid(x,x)

Fxy = (X-h)**2 + (Y-k)**2 - r**2
```

```python
plt.contour(X,Y, Fxy, 0);
plt.plot(h,k,"go")
plt.axis("square");
plt.grid();
plt.axhline(y=0, color="k")
plt.axvline(x=0, color="k");
```

```python
radius_list = np.linspace(0.5,3,10)

h_1 = -1.5
h_2 = 1.5
k = 0

axlim = np.max([abs(h_1),abs(k)]) + np.max(radius_list)
x = np.linspace(-axlim, axlim,100)
X,Y = np.meshgrid(x,x)

for r in radius_list:
    
    Fxy1 = (X-h_1)**2 + (Y-k)**2 - r**2
    Fxy2 = (X-h_2)**2 + (Y-k)**2 - r**2
    
    plt.contour(X,Y, Fxy1, 0, colors=[r/3,r/3,r/3]);  #fractions of rgb
    plt.contour(X,Y, Fxy2, 0,  colors=[r/3,r/3,r/3]);
    
    #plt.plot(h,k,"go")
plt.axis("square");
plt.axis("off");
plt.plot([h_1,h_2],[0,0], color="k", lw=2);
#plt.grid();
    #plt.axhline(y=0, color="k")
    #plt.axvline(x=0, color="k");
    
```

# Graphing Elipses

$$ \frac{(y - k)^2}{b^2} + \frac{(x - h)^2}{a^2} - 1 = 0 $$

```python
h = 1
k = 2

a = 2
b = 3

axlim = np.max([abs(h),abs(k)]) + np.max([a,b])

x = np.linspace(-axlim, axlim, 100)
X,Y = np.meshgrid(x,x)

ellipse = (X-h)**2/(a**2) + (Y-k)**2/b**2 - 1

plt.contour(X,Y,ellipse, 0);
plt.axis("square");
plt.plot(h,k,"go");
plt.axhline(y=0,color="k");
plt.axvline(x=0,color="k");
```

# Graphing Hyperbolas



$$ \frac{(y - k)^2}{b^2} - \frac{(x - h)^2}{a^2} - 1 = 0 $$
$$ Center: (h,k) $$
$$ Radii : a , b $$

```python
a = 1
b = .5
h = 1
k = 2

axlim = 4*np.max([a,b]) + np.max([abs(h),abs(k)])
x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)


# Create the function

hyperbola = (X-h)**2/a**2 - (Y-k)**2/b**2 - 1

plt.contour(X,Y,hyperbola,0);
plt.plot(h,k,"go")
plt.axhline(y=0,color="k", ls="--")
plt.axvline(x=0,color="k", ls="--")
plt.axis("square");
```

```python
n = 16

a = np.linspace(1,5, n)
b = np.linspace(1,5, n)
h = 0
k = 0

axlim = 2*np.max([a,b]) + np.max([abs(h),abs(k)])
x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)

for i in range(0,n):
    
    hyperbola = (X-h)**2/a[i]**2 - (Y-k)**2/b[i]**2 - 1
    plt.contour(X,Y,hyperbola,0, colors=[ 0 ,i/n, 0]);
    plt.contour(Y,X,hyperbola,0, colors=[i/n, 0, i/n]);
    
plt.axis("square")
plt.axis("off");
    
```
