**Asteroid Radian Curve**

$$ x = a * cos(t)^3 $$
$$ y = a * sin(t)^3 $$

$$ \text{For now}\ a = t^1 $$

```python
t = np.linspace(-6*np.pi, 6*np.pi,1000)

a = t

x = a * np.cos(t)**3
y = a * np.sin(t)**3

plt.plot(x,y, lw=2, color="green");
plt.axis("off")
plt.axis("square");
```

now we will use $$ a = t^n $$

```python
t = np.linspace(-6*np.pi, 6*np.pi,1000)

plt.figure(figsize=(15,6),dpi=100)
for n in range(0,9):
    plt.subplot(3,3,n+1)
    

    a = t**n
    x = a * np.cos(t)**3
    y = a * np.sin(t)**3

    plt.plot(x,y, lw=2, color="green");
    plt.axis("off")
    plt.axis("square");
    
```

**Rose Curves**

$$ x = cos(kt) * cos(t) $$
$$ y = cos(kt) * sin(t) $$

$$ \text{if k} \in \mathbb{R}
    \begin{cases}
      \text{k is even}, & \text{2k pedals} \\
      \text{k is odd}, & \text{k pedals}
    \end{cases} $$

```python
k = 7.5

# if k%2 == 0:
#     k = k#k*2
# else:
#     k = k

t = np.linspace(0,4*np.pi,1000)

x = np.cos(k*t) * np.cos(t)
y = np.cos(k*t) * np.sin(t)

plt.plot(x,y,color="red", lw=2)
plt.axis("square")
plt.axis("off");
```

```python
t = np.linspace(0,4*np.pi,1000)

plt.figure(figsize=(15,6),dpi=100)
for n in range(0,9):
    plt.subplot(3,3,n+1)
    
    k = n/2
    x = np.cos(k*t) * np.cos(t)
    y = np.cos(k*t) * np.sin(t)

    plt.plot(x,y, lw=2, color="green");
    plt.axis("off")
    plt.axis("square");
```

**Squircle**

```python
# x**4 + y**4 = 1**4

x = np.linspace(-1,1,2001)
y = (1 - x**4)**(1/4) # this has 2 solutions +-

plt.plot(x,y,"k",lw=2)
plt.plot(x,-y,"k",lw=2)
plt.axis("off")
plt.axis("square");
```

**Logarithmic spirals in nature**

$$ x = cos(t) * e^{t/k} $$
$$ y = sin(t) * e^{t/k} $$

```python
import matplotlib.patheffects as pe

t = np.linspace(-4*np.pi,4*np.pi,1000)



k = -3
x = np.cos(t) * np.exp(t/k) 
y = np.sin(t) * np.exp(t/k) 

plt.plot(x,y,lw=5, color="black", path_effects=[pe.SimpleLineShadow(shadow_color='g', alpha=0.5)]);
plt.axis("off")
plt.axis("square");
```

**Logistic Map**

$$ x_{n+1} = rx_{n}(1 - x_{n}) $$


```python
N = 500
rs = np.linspace(1,4,1000)
endcap = np.arange(round(N*.9),N)


plt.figure(figsize=(10,6),dpi=100)

for r in range(len(rs)):
    x = np.ones(N)/2
    
    for n in range(N-1):
        x[n+1] = rs[r]*x[n]*(1-x[n])
        
    u = np.unique(x[endcap])
    r = rs[r]*np.ones(len(u))
    plt.plot(r,u,'.',markersize=5)#color=[(np.sin(r/len(rs)/2)+1)/2,1-ri/len(rs),.5])

plt.show()
```
