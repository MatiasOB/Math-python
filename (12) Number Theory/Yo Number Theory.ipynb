# <font color="red">**Counting Perfect Numbers**</font>

Number theory is the branch of mathematics that studies the properties of numbers. It is a broad field of study, with its main focus being on natural numbers - numbers like 1, 2, 3, 4, etc. It also includes topics such as modular arithmetic, fractions, prime numbers, and irrational numbers. Number theory has applications in many areas of mathematics, including cryptography, geometry, elementary number theory, and algebraic number theory.

A perfect number is an integer which is the sum of all its positive divisors, excluding the number itself. For example: 6 is a perfect number, because its divisors 1, 2, and 3 add up to 6.

```python
def is_perfect_number(n):

    # Initialize sum
    sum = 0

    # Find divisors and add them to sum
    for x in range(1, n):
        if n % x == 0:
            sum += x

    # Return true if sum of divisors is equal to given number
    return n == sum

x = np.arange(2, 10001)
bool_array = np.vectorize(is_perfect_number)(x)
x[bool_array]
```

# <font color="red">**Euclid's Pythagorean triplets**</font>

In mathematics, a **Pythagorean triple** consists of three positive integers a, b, and c, such that they satisfy the equation: `a^2 + b^2 = c^2`.

The triple is commonly written `(a, b, c)`, and a well-known example is `(3, 4, 5)`, since `3^2 + 4^2 = 9 + 16 = 25 = 5^2`.

```python
x = np.arange(2, 51)


def euclids_formula(m, n):

    if m > n:

        ax = m**2 - n**2
        bx = 2 * m * n
        cx = m**2 + n**2

    else:
        ax = n**2 - m**2
        bx = 2 * m * n
        cx = n**2 + m**2

    return ax, bx, cx


a = []
b = []
c = []

for m in x:
    for n in x:  #for n in range(x[0],m):
        ax, bx, cx = euclids_formula(m, n)
        a.append(ax)
        b.append(bx)
        c.append(cx)
```

```python
sns.scatterplot(x=a,y=b, hue=b, palette="viridis", alpha=0.5 , lw=0.5, edgecolor="k", legend=False)
plt.axis("off");
```

# <font color="red">**Fermat's Theorem**</font>

$$ \huge{x^n + y^n = z^n} $$

Fermats Last Theorem is a concept which was proposed by Pierre de Fermat, stating that there does not exist any natural number _n_ higher than 2 for which the equation described above has any solution. Although there have been attempts to disprove this conjecture, it's been accepted as true since Andrew Wiles proved it in 1995.

```python
ints = np.arange(1,101)

#xx = np.meshgrid(x,x)

Z = np.zeros((len(ints),len(ints)))

for x in ints:
    for y in ints:
        Z[x-1,y-1] = (x**3 + y**3)**(1/3)


# we can test if a number is an integer using number%1 == 0
Zints = Z%1 == 0
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,6))

ax[0].imshow(Z)
ax[0].set_xlabel("y")
ax[0].set_ylabel("x")
ax[0].set_title("$z = \sqrt[3]{x^3+y^3}$", fontsize=16)

ax[1].imshow(Zints)
ax[1].set_xlabel("y")
ax[1].set_ylabel("x")
ax[1].set_title("Whether $z$ is an int", fontsize=16);
```

**Divergent and convergent sequences are two types of number sequences.**

A sequence is said to be convergent if its terms approach a finite limit as the sequence progresses towards infinity. In other words, the sequence gets closer and closer to a particular value as you go further along the sequence. For example, the sequence 1, 1/2, 1/4, 1/8, 1/16, ... is convergent, and its limit is zero.

On the other hand, a sequence is said to be divergent if it does not approach any finite limit as the sequence progresses towards infinity. In other words, the sequence does not converge to a particular value. For example, the sequence 1, 2, 3, 4, 5, ... is divergent, and it continues to increase indefinitely without approaching any finite limit.

It's worth noting that some sequences may not be convergent or divergent in the traditional sense. For example, some sequences oscillate back and forth between two values and do not approach a limit, while others may not have a clear pattern and can exhibit complex behavior.

# <font color="red">**Heron's method of square roots**</font>

$$ \large{\sqrt{Y} = X \rightarrow Y = X^2} $$

$$ Y = (x + \epsilon)^2 $$

$$ \text{Small x will iteratively get closer to big X, and epsilon (error) will get smaller} $$
$$ \text{I want to isolate }\epsilon $$

$$  Y = (x + \epsilon)^2 = (x + \epsilon)(x + \epsilon) = x^2 + \epsilon^2 + 2x\epsilon $$
$$ Y - x^2 = \epsilon^2 + 2x\epsilon = (\epsilon + 2x)\epsilon $$

$$ \text{We assume the }\epsilon \text{ inside the parenthesis is to small when compared to 2x, so we ignore it} $$

$$ \frac{Y-x^2}{2x} \approx \epsilon $$

--------------------------------------------------------------------------------------------------------------------

$$\text{Now considering this 2 equations}$$

$$ Y = (x + \epsilon)^2 \text{ and }\frac{Y-x^2}{2x} \approx \epsilon $$

$$ \text{we replace }\epsilon\text{ in the first equation with the result of the second equation} $$

$$ \text{We get } X \approx x + \frac{Y - x^2}{2x} $$

$$ \text{The algorithm works by iterating big X turns into small x at the next iteration} $$
$$ x \leftarrow x +  \frac{Y - x^2}{2x} $$

$$ \text{we can simplify this iteration math in the following way} $$

$$ x \leftarrow \frac{2x^2}{2x} + \frac{Y-x^2}{2x} $$
$$ x \leftarrow \frac{x^2 + Y}{2x} $$

$$ x \leftarrow \frac{x + \frac{Y}{x}}{2} $$

```python
# define the numbers to compute
Y = np.linspace(2, 101, 50)  # i want to find the sqrt of this numbers

# range of algorithm iterations
n_iter = np.arange(3, 9)


# initialize matrix of estimates
error = np.zeros((len(n_iter), len(Y)))


# loop over the numbers to compute
for n, num in enumerate(Y):

    # loop over number of iterations
    for i, iters in enumerate(n_iter):

        # initial guess
        x = num / 3

        # implement algorithm
        for a in range(iters):
            x = (x + num / x) / 2

        # store error magnitude
        error[i, n] = abs(x - np.sqrt(num))
```

```python
def heron_square_method(num,n_iter):

    # initial random guess
    x = num / 3

    for iter in range(1,n_iter):
        x = (x + num / x) / 2

    return x
```

# <font color="red">**Smooth numbers**</font>

Smooth numbers are positive integers that have only small prime factors. More precisely, a positive integer n is said to be "k-smooth" if its prime factorization contains only primes less than or equal to k.

For example, a number that is 5-smooth (also called "regular") has only prime factors less than or equal to 5. So, the first few 5-smooth numbers are:

1, 2, 3, 4, 5, 6, 8, 9, 10, 12, 15, 16, 18, 20, 24, 25, 27, 30, 32, 36, ...

As you can see, all of these numbers have only prime factors less than or equal to 5. The number 7, for example, is not 5-smooth because it has a prime factor greater than 5 (namely, 7 itself).

Smooth numbers are important in number theory and have applications in cryptography and factorization algorithms, among other areas.

```python
# largest number to search in
maxN = 10000

# find the largest prime factor
largestPrimeFact = np.zeros(maxN+1,dtype=int)

for i in range(2,maxN+1): # note: start at 2! not zero!
    largestPrimeFact[i] = np.max(sym.primefactors(i))

# show the smooth numbers on a plot
fig,ax = plt.subplots(1,figsize=(10,7))
plt.plot(largestPrimeFact,'k.',markersize=.5)
plt.xlabel('Number')
plt.ylabel('Largest prime factor')
plt.grid(True)
plt.show()
```

```python
# count the number of occurrences of maximum prime factors

# find all unique max primes (basically all the primes)
uniqueMaxPrimes = np.unique(largestPrimeFact)

print(largestPrimeFact.shape)
print(uniqueMaxPrimes.shape)

# count the number of times each max-prime appears
num = np.zeros(len(uniqueMaxPrimes), dtype=int)
for i, u in enumerate(uniqueMaxPrimes):
    num[i] = np.sum(largestPrimeFact == u)

print(num.shape)

# and visualize the counts by the smooth numbers
plt.plot(uniqueMaxPrimes, num, "k.")
plt.xlabel("Max prime number")
plt.ylabel("Number of occurrences")
plt.xlim([-10, 500])
# plt.ylim([2,300])
plt.title("%s is the modal max prime factor" % uniqueMaxPrimes[np.argmax(num)])
plt.show()
```

Find integer solutions to $x^1+y^2=z^3$

```python
ints = np.arange(1,101)

z = np.zeros((len(ints),len(ints)))

# run the simulation
for x in ints:
    for y in ints:
        z[x-1,y-1] = (x**1 + y**2)**(1/3)

# check for integers
zInts = z%1 == 0

# now visualize
fig,ax = plt.subplots(1,2,figsize=(12,6))
fig.tight_layout(pad=5)

# show the resulting z-values
h = ax[0].imshow(z)
fig.colorbar(h,ax=ax[0],fraction=.45)
ax[0].set_xlabel('y')
ax[0].set_ylabel('x')
ax[0].set_title('$z=\sqrt[3]{x^1+y^2}$',fontsize=16)

# show the boolean integer map
h = ax[1].imshow(zInts,vmin=0,vmax=1)
fig.colorbar(h,ax=ax[1],fraction=.045)
ax[1].set_xlabel('y')
ax[1].set_ylabel('x')
ax[1].set_title('Whether $z$ is int',fontsize=12)

plt.show()
```

Compute and plot the following two sequences:

$\huge 
s_1 = \frac{-1^a}{\ln(\frac{1}{e^a})}$

$\huge
s_2 = \frac{\sqrt{a}}{\left(\frac{-\pi}{2}\right)^a}$

for $a \in \mathbb{Z}$

```python
a = np.arange(1,21)

s1 = np.zeros(len(a))
s2 = np.zeros(len(a))

for a in a:
    s1[a-1] = (-1)**a / np.log(1/np.exp(a))
    s2[a-1] = np.sqrt(a) / ((-np.pi)/2)**a


fig,ax = plt.subplots(1,2,figsize=(10,4))
ax[0].plot(s1)
ax[0].set_title('Sequence $s_1$')

ax[1].plot(s2)
ax[1].set_title('Sequence $s_2$')

plt.show()
```
