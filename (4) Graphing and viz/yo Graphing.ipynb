```python
x = sym.symbols("x")

f1 = x**2 - 3*x

xi = np.arange(-10,11)
yi = [f1.subs({x:n}) for n in xi]

display(f1.factor())
plt.plot(xi,yi, "ro");
#plt.axis("square");
plt.grid();
```

```python
x = sym.symbols("x")

f1 = (x**2 - 3*x)

xi = np.arange(-7,11)
yi = [f1.subs({x:n}) for n in xi]

display(f1.factor())

for i, n in enumerate(sym.solve(f1,x)):
    display(Math(sym.latex(f"x{i+1} = ") + sym.latex(n)))
plt.plot(xi,yi, color=[.4,.2,.7], lw=3);
plt.plot(np.arange(-20,20), [0] * len(np.arange(-20,20)), color="black")
plt.plot([0] * len(np.arange(-20,200)), np.arange(-20,200), color="black")

plt.ylim([-5,70])
plt.xlim([-7,10])
#plt.axis("square");
plt.grid();
```

```python
x = sym.symbols("x")

f1 = sym.root(np.abs(x),2)

xi = np.arange(-20,21)
yi = [f1.subs({x:n}) for n in xi]


fig, ax = plt.subplots(figsize=(10,6))
for n in xi:
    ax.plot([0,n], [0,f1.subs({x:n})]);
    ax.plot(n, f1.subs({x:n}), "o");
    ax.plot(n, f1.subs({x:n})*-1,"o");
    ax.plot([0,n], [0,f1.subs({x:n})*-1])
```

```python
import sympy.plotting.plot as symplot

x, a  = sym.symbols("x")

y = x**2

p = symplot(y,show=False)
p.xlim = [0,50]
p[0].line_color = "m"
p.title = "This is a nice plot!"
print(p)
p.show()
```

```python
x, a  = sym.symbols("x,a")

expr = a/x

n=2

p = symplot(expr.subs({a:n}), (x,-5,5), show=False)
p.ylim = [-5,5]
p[0].label = f"$y = %s$" %(sym.latex(expr.subs({a:n})))

n=1

p2 = symplot(expr.subs({a:n}), (x,-5,5), show=False)
p.extend(p2)
p[1].label = f"$y = %s$" %(sym.latex(expr.subs({a:n})))


p.legend = True
p.show()
```

```python
x, a  = sym.symbols("x,a")

expr = a/(x**2-a)

p = symplot(expr.subs({a:1}), show=False)
p.ylim = [-10,10]
p.xlim = [-5,5]
p[0].label = f"$y = %s$" %(sym.latex(expr.subs({a:1})))

for n in range(2,5):

    p1 = symplot(expr.subs({a:n}), show=False)
    p.extend(p1)
    p[n-1].label = f"$y = %s$" %(sym.latex(expr.subs({a:n})))
    p[n-1].line_color = np.random.rand(3)
    
    
p.legend = True
p.show()
```

```python
a = np.zeros((10,14),dtype="int")

for row in range(0,a.shape[0]):
    for column in range(0,a.shape[1]):
        a[row,column] = 3*row - 4*column

display(sym.sympify(a))
plt.imshow(a);
plt.plot([0,3],[8,2],"r",lw=4);

for row in range(0,a.shape[0]):
    for column in range(0,a.shape[1]):
        plt.text(column, row , a[row,column], horizontalalignment="center", verticalalignment="center");
        
plt.set_cmap("Purples")
```
